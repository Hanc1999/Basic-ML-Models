## 4. Cross-Validation(5) + SVM

### 4.1 CV function

```python
from sklearn.svm import LinearSVC
from sklearn.datasets import make_classification

# define cross validation function differ with cost
def cross_validation_5(cost):
    wrong = 0
    for i in range(1,6):
    	# split dataset
        test_X = X_data[(i-1)*200:i*200]
        test_y = y_data[(i-1)*200:i*200]
        train_X = np.delete(X_data, [a for a in range((i-1)*200, i*200)], axis=0)
        train_y = np.delete(y_data, [b for b in range((i-1)*200, i*200)], axis=0)

        # train the SVM model
        clf = LinearSVC(random_state=0, tol = cost, max_iter=10000)
        clf.fit(train_X, train_y) 

        predict = clf.predict(test_X)
        for k in range(len(predict)):
            if predict[k] != test_y[k]:
                wrong += 1
                
    return wrong / 1000
```

```python
# use cv5 function with different cost

cost = []
error_rate = []
mis_num = []
i = 3.16
while i <= 3.22:
    i = float(format(i, '0.3f'))
    cost.append(format(i, '0.3f'))
    error_rate.append(cross_validation_5(i))
    mis_num.append(int(cross_validation_5(i)*1000))
    i += 0.001
    
```

```python
# train-test result table:
tb = pt.PrettyTable()
tb.add_column("Cost",cost)
tb.add_column("Error Rate",error_rate)
tb.add_column("Misclassified Number", mis_num)
print('Table for Error Rate with Different Cost Value')
print(tb,'\n')
```

### 4.2 Test SVM model

```python
#test data set
Test_X, Test_y = make_classification(n_samples=1000, # 样本个数
n_features=2, # 特征个数
n_informative=2, # 有效特征个数
n_redundant=0, # 冗余特征个数（有效特征的随机组合）
n_repeated=0, # 重复特征个数（有效特征和冗余特征的随机组合）
random_state=0)

# plot test data
x_0 = []
y_0 = []
x_1 = []
y_1 = []
for j in range(len(Test_X)):
    if Test_y[j] == 1:
        x_1.append(Test_X[j][0])
        y_1.append(Test_X[j][1])
    else:
        x_0.append(Test_X[j][0])
        y_0.append(Test_X[j][1])

plt.scatter(x_0, y_0, color='navy', marker='.', label='0')
plt.scatter(x_1, y_1, color='green', marker='x', label='1')
plt.title('Test Data Set')
plt.legend()
plt.show()

```

```python
# compute test error with different cost value
# find two regions where misclassified number is 1 (at minimum)
cost = []
error_rate = []
mis_num = []
i = 1.20
while i <= 1.23:
    i = float(format(i, '0.3f'))
    cost.append(format(i, '0.3f'))
    error_rate.append(cross_validation_5(i))
    mis_num.append(int(cross_validation_5(i)*1000))
    i += 0.001

# draw the table
tb = pt.PrettyTable()
tb.add_column("Cost",cost)
tb.add_column("Error Rate",error_rate)
tb.add_column("Misclassified Number", mis_num)
print('Table for Test Error Rate with Different Cost Value')
print(tb)
```
